```python
import requests

url = "https://vegetablemarketprice.com/market/karnataka/today#google_vignette"
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully retrieved the webpage!")
    html = response.text
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
from bs4 import BeautifulSoup

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Print the parsed HTML (optional, for inspection)
print(soup.prettify())
```

```python
table_body = soup.find('tbody', id='todayVegetableTableContainer')
vegetable_data = []
```

```python
if table_body:
    # Find all vegetable rows in the table
    rows = table_body.find_all('tr', class_='todayVegetableTableRows')
    
    for row in rows:
        # Extract the columns of each row (td tags)
        cols = row.find_all('td')
        
        # Extract relevant information from the row
        if len(cols) > 0:
            vegetable = {
                'Vegetable': cols[1].get_text(strip=True),  # Vegetable Name (2nd column)
                'Wholesale Price': cols[2].get_text(strip=True),  # Wholesale Price (3rd column)
                'Retail Price': cols[3].get_text(strip=True),  # Retail Price (4th column)
                'Shopping Mall': cols[4].get_text(strip=True),  # Shopping Mall (5th column)
                'Units': cols[5].get_text(strip=True),  # Units (6th column)
            }
            vegetable_data.append(vegetable)
```

```python
print("Vegetable Data:")
for data in vegetable_data:
    print(data)
```

```python
import csv

# Define the field names (headers) for the CSV file
fieldnames = ['Vegetable', 'Wholesale Price','Retail Price','Shopping Mall','Units']

# Write to a CSV file
with open('vegetable_prices.csv', 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()  # Write the header row
    writer.writerows(vegetable_data)  # Write the data rows

print("Data saved to vegetable_prices.csv")
```

```python
csv_file = 'vegetable_data.csv'  # Specify the name of the CSV file

# Open the CSV file in write mode
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    # Create a CSV writer object
    writer = csv.DictWriter(file, fieldnames=['Vegetable', 'Wholesale Price', 'Retail Price', 'Shopping Mall', 'Units'])
    
    # Write the header row
    writer.writeheader()
    
    # Write the vegetable data rows
    writer.writerows(vegetable_data)

print(f"Data successfully saved to {csv_file}")
```
